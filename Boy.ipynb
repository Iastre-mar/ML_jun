```python
from matplotlib.pyplot import imshow
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
import sys
img = mpimg.imread('Boy.jpg')
print(sys.getsizeof(img))
imshow(img)
plt.show()
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
def rgb2gray(rgb):
    tile = np.tile(np.c_[0.333, 0.333, 0.333], reps=(rgb.shape[0],rgb.shape[1],1))
    return np.sum(tile * rgb, axis=2)

img_gray = rgb2gray(img)
print(type(img_gray), img_gray.shape)
imshow(img_gray, cmap = "gray")
plt.show()
```

```python
svd_model = TruncatedSVD(n_components=30).fit(img_gray)
X_svd = svd_model.transform(img_gray)
print(sys.getsizeof(X_svd))
X_svd_restored = svd_model.inverse_transform(X_svd)
imshow(X_svd_restored, cmap = "gray")
print(sys.getsizeof(X_svd_restored))
plt.show()
plt.savefig('Boy2.jpg')
```
